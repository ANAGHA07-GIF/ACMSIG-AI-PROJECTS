```python
class KNN:
  def __init__(self, n_neighbors=7):             #Assign the number of nearest neighbours .Preferably an odd small number (5,7,or 9)
        self.n_neighbors = n_neighbors

  def euclidean_distance(self, x1, x2):          #Calculate the euclidean distance the distance between the training data and the nearest neighbours.Can use other methods to calculate the ares as well.
      return np.linalg.norm(x1 - x2)
  def fit(self, X_train, y_train):               #
    self.X_train = X_train
    self.y_train = y_train
  def predict(self, X):
    predictions = []
    for x in X:
      prediction = self._predict(x)
      predictions.append(prediction)
    return np.array(predictions)
  def _predict(self, x):
    distances = []
    for x_train in self.X_train:
            distance = self.euclidean_distance(x, x_train)
            distances.append(distance)
    distances = np.array(distances)
    n_neighbors_idxs = np.argsort(distances)[: self.n_neighbors]
    labels = self.y_train[n_neighbors_idxs]
    labels = list(labels)
    most_occuring_value = max(labels, key=labels.count)
    return most_occuring_value
```

```python
def train_test_split(X, y, random_state=42, test_size=0.2):
  n_samples = X.shape[0]
  np.random.seed(random_state)
  shuffled_indices = np.random.permutation(np.arange(n_samples))
  test_size = int(n_samples * test_size)
  test_indices = shuffled_indices[:test_size]
  train_indices = shuffled_indices[test_size:]
  X_train, X_test = X[train_indices], X[test_indices]
  y_train, y_test = y[train_indices], y[test_indices]

  return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size = 0.2, random_state=42)
```

```python
model = KNN(7)
model.fit(X_train, y_train)
```

```python
def compute_accuracy(y_true, y_pred):
    y_true = y_true.flatten()
    total_samples = len(y_true)
    correct_predictions = np.sum(y_true == y_pred)
    return (correct_predictions / total_samples)
```

```python
predictions = model.predict(X_test)
accuracy = compute_accuracy(y_test, predictions)
print(f" our model got accuracy score of : {accuracy}")
```
